```python
import pandas as pd
import numpy as np
import random
from sklearn.metrics.pairwise import euclidean_distances
```

```python

def createPreferences(experts_number,number_of_alternatives, probCertainty,probConsistency,NormalUncertainty):
    #i
    i_numbers = (number_of_alternatives-1)
    #t_and_f
    t_and_f_numbers = (number_of_alternatives*number_of_alternatives) - number_of_alternatives
    
    preferences= pd.DataFrame(np.zeros((experts_number,t_and_f_numbers),dtype=object))
    indeterminacy= pd.DataFrame(np.zeros((experts_number,i_numbers),dtype=object))

    for num in range(experts_number):
        if(np.random.negative_binomial(1, 1-probConsistency, 1)):
            preferences.loc[num] = generateConsistentPref(t_and_f_numbers)
        else:
            preferences.loc[num] = generatePref(t_and_f_numbers)                  
     
        if (NormalIndeterminacy):
            for i in range(int(i_numbers)):
                s = np.random.normal(0.5, 0.1, experts_number)
                indeterminacy[i]=s
            break 
            
        else:
            indeter=[]    
            for i in range(int(i_numbers)): 
                if(np.random.negative_binomial(1, 1-probCertainty, 1)):       
                    indeter.append(generateCertainI(probCertainty))
                else:
                    indeter.append(generateI())
            indeterminacy.loc[num] = indeter  
               
    preferences.to_csv('preferences'+str(experts_number)+'-'+str(number_of_alternatives)+'.csv', encoding='utf-8', index=False, header=False)
    indeterminacy.to_csv('indeterminacies'+str(experts_number)+'-'+str(number_of_alternatives)+'.csv', encoding='utf-8', index=False)
    return (indeterminacy,preferences)
    
    
def generateConsistentPref(t_and_f_numbers):# Generate fully consistent alternatives
    pref= np.zeros((t_and_f_numbers),dtype=object)
    for j in range(int(t_and_f_numbers/2)):
        pref[j]=random.uniform(0,1)
        pref[j+(int(t_and_f_numbers/2))]=1-pref[j]          
    return (pref)
    
def generatePref(t_and_f_numbers):
    pref= np.zeros((t_and_f_numbers),dtype=object)
    for j in range(t_and_f_numbers):
        pref[j]=random.uniform(0,1)  
    return (pref)

def generateCertainI(probCertainty):# Generate Certainty
    return np.random.choice([0,1])
       
        
def generateI():
    return np.random.random(1)
```

```python
experts_number = 50# Number of experts
number_of_alternatives = 3 # Number of alternatives
NormalIndeterminacy=False #the distribution is not normal
probCertainty=0.1  #probility of determine preferenced
probConsistency =0.2 #probility of consistent preferenced
indeterminacy,preferences = createPreferences(experts_number, number_of_alternatives, probCertainty,probConsistency,NormalIndeterminacy)
```
